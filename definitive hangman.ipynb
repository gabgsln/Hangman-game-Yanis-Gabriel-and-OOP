```python
import random
def is_length_correct(word, length):
    return len(word) == length

def all_words(length):#we will like sort the worrds and distinguish them in function of their lenght#
    all_words = []
    with open("words.txt") as f:
        for line in f: 
            line = line.strip()
            if is_length_correct(line, length):
                all_words.append(line)
    return all_words
  
def random_select(words_list):
    return random.choice(words_list) #here the computer choose a word for the selected length#

def show_hangman(errors):#list of shcemss of each erros and as the errrors increase the indice of the list stages increase and giveaway a evolving shceme#
    stages = [
        """
           --------
           |      |
           |      
           |    
           |      
           |     
           -
        """,
        """
           -------- 
           |      |
           |      O
           |    
           |      
           |     
           -
        """,
        """
           --------
           |      |
           |      O
           |      |
           |      |
           |     
           -
        """,
        """
           --------
           |      |
           |      O
           |     \|
           |      |
           |     
           -
        """,
        """
           --------
           |      |
           |      O
           |     \|/
           |      |
           |     
           -
        """,
        """
           --------
           |      |
           |      O
           |     \|/
           |      |
           |     / 
           -
        """,
        """
           --------
           |      |
           |      O
           |     \|/
           |      |
           |     / /
           -
        """
    ]
    return stages[errors]

def play_game():
    guessed = False
    score = 0
    guessed_letters = []
    guessed_words = []
    length = int(input("Choose the length of the word you want: "))
    all_words_list = all_words(length)
    word = random_select(all_words_list).upper()
    unguessed_word = "_" * len(word)
    errors = 0
    max_errors = 6

    print("Welcome to Hangman game!")
    print(unguessed_word)
    print(f"You have {max_errors} triess.")

    while not guessed and errors!=max_errors:
        guess = input("Guess a letter or the full word: ").upper()

        if len(guess) == 1:  # Guess a letter for len 1 so for a letter#
            if guess in guessed_letters:
                print(f"You already guessed the letter {guess}.")
            elif guess not in word:
                print(f"The letter {guess} is not in the word.")
                errors += 1
                score -= 5
                guessed_letters.append(guess)
            else:
                print(f"Good job! {guess} is in the word.")
                guessed_letters.append(guess)
                word_as_list = list(unguessed_word)
                indices = [i for i, letter in enumerate(word) if letter == guess]
                for index in indices:
                    word_as_list[index] = guess
                unguessed_word = "".join(word_as_list)

                if "_" not in unguessed_word:  #when there will be no more _ its mean the word will be completely guessed
                    guessed = True

        elif len(guess) == len(word):  # if you want to guess and all word #
            if guess in guessed_words:
                print("This word was already guessed previously.")
            elif guess != word:
                print(f"{guess} is not the word.")
                errors += 1
                score -= 5
                guessed_words.append(guess)
            else:
                guessed = True
                unguessed_word = word
                score += 50

        else:
            print("Not a valid guess")

        print(show_hangman(errors))
        print(unguessed_word)

    if guessed:
        print("You won")
    else:
        print(f"Sorry, you lost. The word was {word}. Your score is {score}.")

    play_again()

def play_again():
    response=False
    response = input("Do you want to play again? (yes/no): ").lower()
    if response == "yes":
        response=True
        play_game()
    else:
        print("Thanks you for playing")

if __name__ == "__main__":
    play_game()



```
